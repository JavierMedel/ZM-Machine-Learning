## Matplotlib example workflow

```python
# 0.- import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.- prepare data
x = [1,2,3,4]
y = [5,6,7,8]

# 2.- Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # widht, hight

# 3.- Plot data
ax.plot(x,y)

# 4.- Customize plots
ax.set(title="Simple Plot",
       xlabel='x-axis',
       ylabel='y-axis')

# 5.- Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

## Two subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize = (10, 5))
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100))
plt.show()
```

```python
# Subplots options 2
fig, ax = plt.subplots(nrows=2, 
                       ncols=2, 
                       figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

# Which one should I use? (pyplot vs matplotlib OO Method?)

* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advance, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50.head())
```

```python
#Pylot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
plt.show()
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
ax.set_xlim([45,100])
plt.show()
```

```python
over_50['target'].values
```

```python
# OO Mething from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50['chol'].mean(), 
           linestyle='--')

plt.show()
```

```python
heart_disease.head()
```

```python
over_50.head()
```

```python
# Subplot of c
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10,10), 
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target'])
# Customize ax0
ax0.set(title='Heart and Cholesterol Levels', 
        ylabel='Cholesterol')


# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), 
           linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')


plt.show()
```

```python
# set the style
plt.style.use('seaborn-whitegrid')

# Subplot of c
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10,10), 
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target'], 
                       cmap='plasma') # this change the colour scheme
# Customize ax0
ax0.set(title='Heart and Cholesterol Levels', 
        ylabel='Cholesterol')


# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), 
           linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'], 
                      cmap='summer')
# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')


plt.show()
```

```python
# Subplot of c
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10,10), 
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target'], cmap='winter')
# Customize ax0
ax0.set(title='Heart and Cholesterol Levels', 
        ylabel='Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), 
           linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'], 
                      cmap='winter')
# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate', 
        xlabel='Age', 
        ylabel='Max Heart Rate')

# Change the x axis limit
ax1.set_xlim([50,80])
# Change the y axis limit
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')


plt.show()
```
