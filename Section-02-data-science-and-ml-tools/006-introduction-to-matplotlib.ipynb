## Matplotlib example workflow

```python
# 0.- import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.- prepare data
x = [1,2,3,4]
y = [5,6,7,8]

# 2.- Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # widht, hight

# 3.- Plot data
ax.plot(x,y)

# 4.- Customize plots
ax.set(title="Simple Plot",
       xlabel='x-axis',
       ylabel='y-axis')

# 5.- Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(3,3))
ax.plot(x, x**2)
plt.show()
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# another scatter plot
fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary

nut_butter_prices = {"Almond butter" : 10,
                     "Peanut butter" : 8,
                     "Cashew butter" : 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) #x, y
ax.set(title = 'Butter Store', ylabel='Price ($)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize = (10, 5))
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100))
plt.show()
```

```python
# Subplots options 2
fig, ax = plt.subplots(nrows=2, 
                       ncols=2, 
                       figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

# Which one should I use? (pyplot vs matplotlib OO Method?)

* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advance, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50.head())
```

```python
#Pylot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
plt.show()
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
ax.set_xlim([45,100])
plt.show()
```
